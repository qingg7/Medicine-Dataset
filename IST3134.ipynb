# Data Cleaning and Standardization

```python
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from string import punctuation

# Download the 'punkt' tokenizer data (if not already downloaded)
nltk.download('punkt')
nltk.download('stopwords')

# Convert text data to lowercase
data_frame['DrugName'] = data_frame['DrugName'].str.lower()
data_frame['Condition'] = data_frame['Condition'].str.lower()
data_frame['Review'] = data_frame['Review'].str.lower()

# Define a function to remove punctuation, tokenize, remove stopwords, and apply stemming
def preprocess_text(text):
    # Remove punctuation (including slash, hashtags, and numbers)
    text = re.sub(r'[{}]|\d'.format(punctuation), '', text)
    
    # Tokenize into individual words
    words = nltk.word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Apply stemming
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]
    
    return words

# Apply preprocessing to 'DrugName', 'Condition', and 'Review' attributes
data_frame['DrugName'] = data_frame['DrugName'].apply(preprocess_text)
data_frame['Condition'] = data_frame['Condition'].apply(preprocess_text)
data_frame['Review'] = data_frame['Review'].apply(preprocess_text)

# Display the DataFrame after preprocessing
display(data_frame)
```

# Generate Wordclouds

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to perform Word Count on the 'Review' attribute
def word_count(data_frame):
    word_count_dict = {}
    for review in data_frame['DrugName']:
        words = review.split()
        for word in words:
            word = word.lower()
            word = word.strip('.,!?()[]{}":;')
            if word not in word_count_dict:
                word_count_dict[word] = 1
            else:
                word_count_dict[word] += 1
    return word_count_dict

# Function to create and display the Word Cloud
def create_word_cloud(word_count_result):
    max_word_count = max(word_count_result.values())
    wordcloud = WordCloud(width=800, height=400, background_color='white',
                          max_font_size=100, colormap='viridis',
                          prefer_horizontal=0.7, max_words=200).generate_from_frequencies(word_count_result)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Load the CSV file into a DataFrame
file_path = 'C:\\Users\\yiqin\\Downloads\\Sem 9.csv'
data_frame = pd.read_csv(file_path)

# Perform Word Count on the 'Review' attribute
word_count_result = word_count(data_frame)

# Display the word count results
for word, count in word_count_result.items():
    print(f"{word}: {count}")

# Create and display the Word Cloud with custom sizes and colors
create_word_cloud(word_count_result)
```
